```python
import matplotlib.pyplot as plt

# Load file
stat_file = open('./Stats/forcedpath.txt', 'r')
forced_frac = stat_file.readlines()
forced_frac = [float(f.strip()) for f in forced_frac]

# Plot
plt.figure(figsize=(8, 6))
plt.hist(forced_frac)
plt.xlabel('Fraction of forced transitions')
plt.ylabel('Number of occurences')
plt.show()
```

```python
# Load file
stat_file = open('./Stats/forcedvssize.txt', 'r')
label = []
forced = []
size = []
for line in stat_file:
    data = line.strip().split(' ')
    label.append(data[0])
    forced.append(float(data[1]))
    size.append(float(data[2]))

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(size, forced)
plt.xscale('log')
plt.xlabel('Data size')
plt.ylabel('Average fraction of forced transitions')
for i, l in enumerate(label):
    plt.annotate(l, (size[i], forced[i]))
plt.show()
```
